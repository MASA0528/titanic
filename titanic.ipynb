```python
# 'Age'を補完
for i in range(len(X)):
    if pd.isna(X.loc[i, 'Age']):
        if 'Mr.' in X.loc[i, 'Name']:
            mean_age = X.loc[X['Name'].str.contains('Mr.', na=False) & X['Age'].notna(), 'Age'].mean()
            X.loc[i, 'Age'] = mean_age

        elif 'Mrs.' in X.loc[i, 'Name']:
            mean_age = X.loc[X['Name'].str.contains('Mrs.', na=False) & X['Age'].notna(), 'Age'].mean()
            X.loc[i, 'Age'] = mean_age

        elif 'Miss.' in X.loc[i, 'Name']:
            mean_age = X.loc[X['Name'].str.contains('Miss.', na=False) & X['Age'].notna(), 'Age'].mean()
            X.loc[i, 'Age'] = mean_age

        else:
            mean_age = X['Age'].mean()
            X.loc[i, 'Age'] = mean_age
```

```python
#欠損地の補完
X['Embarked'] = X['Embarked'].fillna(X['Embarked'].mode()[0])
```

```python
X['FamSize'] = X['SibSp'] + X['Parch']+ 1
```

```python
X['isAlone'] = 0

for i in range(len(X['isAlone'])):
    if X['FamSize'].iloc[i] == 1:
        X['isAlone'].iloc[i] = 1
```
